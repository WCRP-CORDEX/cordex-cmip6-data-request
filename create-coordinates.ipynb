```python
def get_vertical_dimension(df, type="p"):
    dims = df[df["dimensions"].str.contains(type)].dimensions.unique()
    vdims = [d.split(" ")[-1] for d in dims]
    return list(dict.fromkeys(vdims))
```

```python
get_vertical_dimension(df, "height")
```

```python
import re


def get_value_from_out_name(out_name):
    try:
        return re.search(r"\d+", out_name).group()
    except Exception:
        return None
```
